```python
import cv2
```

```python
clicked = False
def onMouse(event, x, y, flags, param):
    global clicked
    if event == cv2.EVENT_LBUTTONUP:
        clicked = True
```

```python
cameraCapture = cv2.VideoCapture(0)
if cameraCapture is None or not cameraCapture.isOpened():
    print('Warning: unable to open video source: 0')
```

```python
cv2.namedWindow('Captured Video Preview')
cv2.setMouseCallback('Captured Video Preview', onMouse)
```

```python
print('Showing camera feed. Click window or press any key to stop.')
success, frame = cameraCapture.read()
while success and cv2.waitKey(1) == -1 and not clicked:
    cv2.imshow('Captured Video Preview', frame)
    success, frame = cameraCapture.read()

cv2.destroyWindow('Captured Video Preview')
cameraCapture.release() 
```
